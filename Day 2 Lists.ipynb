```python
y=[22,22.33,"hello",True,45,2+3j]
y
```

```python
print(y[:5])     #will skip the last element
print(y[0:])     # every element will be there
print(y[1:4])
print(y[3:4])    #positive indexing
print(y[-6:-2])  #negative indexing
print(y[:])      #whole list elements will be there
```

```python
print(y[-2:-4])   
                   #left to right indexing will only work
print(y[4:1])
```
