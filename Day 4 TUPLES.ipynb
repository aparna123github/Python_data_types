
### Tuple Methods

#### tuple have only two methods; count & index

```python
tuple = (2,4,5,8,2,6,9,4,4,6,7,94,2,2)
x = tuple.count(2)
print(x)                 
```

```python
tuple1= ('s','s','h','j','k')
z= tuple1.count('s')
print(z)
```

```python
tuple2 = (83,35,52,11,44,940)
y = tuple2.index(940)
print(y)
```

```python
tuple3 = ('megha', 'palak', 'debo')
z = tuple3.index('debo')
print('my fav person is:',(z))
```
