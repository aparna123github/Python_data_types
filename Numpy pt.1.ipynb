```python
a = np.array([3, 6, 9, 12])
b = a[1:3]
b[:] = 0
print(a)
```

```python
#to avod the prior situation
a = np.array([3, 6, 9, 12])
b = a[1:3].copy()
b[:] = 0
print(a)
```

```python
a = np.array([2, 4, 6, 8, 10])
b = a[2:]
b[0] = 99
print(a)
```

```python
a = np.array([1, 2, 3, 4])
b = a[1:3].copy()
b[0] = 100
print(a)  #copy was there ,therfore it will give the original array
print(b)  #if b is printed seprately then it will work
```
