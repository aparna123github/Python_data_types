### List op prediction ques

```python
a = [10, 20, 30, 40]
print(a[1:3])
# op - 20,30,40
```

```python
a = [1, 2, 3]
a += [4, 5]
print(a)
#op- 1,2,3,4,5
```

```python
a = [1, 2, 3]  #WRONG
a *= 2
print(a)
#op- 2,4,6
```

```python
a = [1, 2, 3]
print(a.pop(), a)
# 1,2
```

```python
a = [3, 1, 2]
a.sort()
print(a)
# 1,2,3
```

```python
a = [1, 2, 3]
print(a[::-1])
#3,2,1
```

```python
a = [1, 2, 3]
del a[1]
print(a)
#2,3
```
