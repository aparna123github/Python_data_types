### Loop control statements
#### pass
#### break
#### continue

```python
count = 5
while count > 0:
   if count ==3:
       pass
   else:
       print(count)
       count -=1
```

```python
for i in range(5):
    if i == 3:
        break
    print(i)
```

```python
for i in range(5):
    if i == 3:
         continue
    print(i)
```

#### Nested Loop

```python
for num in range(1,4):        #PRINTING NO. FROM 1 TO 3 FOR THREE TIMES
    print(num)
for num in range(1,4):
    print(num)
for num in range(1,4):
    print(num)
for num in range(1,4):
    print(num)
for num in range(1,4):       
    print(num)              #TO GET RID OF THIS CONDITION WE USE NESTED LOOPS
```

```python
for i in range(3):           #will do it for 4 times,THIS IS THE OUTER LOOP
    for num in range(1,4):   #for this range*,INNER LOOP
        print(num)
    print("----")            #PRINTING NO. FROM 1 TO 3 FOR THREE TIMES
```
