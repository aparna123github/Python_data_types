### Comparison Operators

### Identity and membership Operator
#### used to compare memory location,not values

```python
a = 10
b = 10
               #IS operator
print(a is b)  #true bcoz both the values are pointing to the same memory location
```

```python
a = 13 
b = 37
print(a is b) #not the same memory location
```

```python
x = [1, 2, 3]
y = [1, 2, 3]
                   #IS NOT operator
print(x is not y)  #same values but different objects in memory
```

```python
x = [1, 2, 3]
y = [1, 2, 3]    ##SAME VALUE DIFFERENT MEMORY
print(x == y)    # True (bcoz the values are equal)
print(x is y)    # False(bcoz two diff lists are stored in different memory location)
```
