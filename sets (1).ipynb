## set operation

### union

```python
s4 = {1,2,3,4,5}
s5 = {4,5,6,7,8}
s6 = {8,9,10}
```

```python
s4.union(s5)
```

```python
s4.union(s6,s5)
```

```python
s5 | s6 | s4 #any order will work
```

### intersection

```python
s4 = {1,2,3,4,5}
s5 = {4,5,6,7,8}
s6 = {8,9,10}
```

```python
s5.intersection(s6)
```

```python
s4 & s5
```

### difference

```python
s4 = {1,2,3,4,5}
s5 = {4,5,6,7,8}
s6 = {8,9,10}
```

```python
s4.difference(s5)
```

```python
s5.difference(s4)
```

```python
s4 - s6   #the diff of the first variable will be shown
```

```python
s5 - s6
```

### symmetric differnce

```python
s4 = {1,2,3,4,5}
s5 = {4,5,6,7,8}
s6 = {8,9,10}
```

```python
s4.symmetric_difference(s6)  #common elemnets will be removed and other elemnts will be shown
```

```python
s6.symmetric_difference(s5)
```

```python
s4 = {1,2,3,4,5}
s5 = {4,5,6,7,8}
s6 = {8,9,10}
print(s4)
print(s5)
print(s6)
```

```python
s4.symmetric_difference_update(s5)  #the variable would be updated and the common value will be elliminated
s4
```

```python
s4.symmetric_difference_update(s6)
s4
```
